```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
import pickle as pkl
import matplotlib.pyplot as plt
from plot_functions import plot_4_classifiers
import warnings
warnings.filterwarnings('ignore')
```

### logistic regression and a support vector machine to classify images of handwritten digits

```python
with open('X.pkl', 'rb') as f:
    X = pkl.load(f)

with open('y.pkl', 'rb') as f:
    y = pkl.load(f)
```

```python
classifiers = [LogisticRegression(), LinearSVC(), SVC(gamma = 'auto'), KNeighborsClassifier()]
```

```python
for c in classifiers:
    c.fit(X, y)

plot_4_classifiers(X, y, classifiers)
plt.show()
```
