```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import normalize
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import pickle as pkl
```

```python
df = pd.read_csv('company_stocks.csv')
```

```python
companies = df.iloc[:, 0].to_list()
prices = np.array(df.iloc[:, 1:])
```

```python
norm_prices = normalize(prices)
```

```python
model = TSNE(learning_rate=50)
```

```python
tsne_features = model.fit_transform(norm_prices)
```

```python
xs = tsne_features[:,0] 
ys = tsne_features[:,1]
```

```python
plt.figure(figsize=(16,8))
plt.scatter(xs, ys, alpha=0.5)
for x, y, company in zip(xs, ys, companies):
    plt.annotate(company, (x, y), fontsize=5, alpha=0.75)
plt.show()
```

```python
with open('variety_numbers.pkl', 'rb') as f:
    variety_numbers = pkl.load(f)
    
with open('samples2.pkl', 'rb') as f:
    samples2 = pkl.load(f)    
```

```python
model = TSNE(learning_rate=200)
tsne_features = model.fit_transform(samples2)
```

```python
xs = tsne_features[:,0]
ys = tsne_features[:,1]
plt.scatter(xs, ys, c=variety_numbers)
plt.show()
```
